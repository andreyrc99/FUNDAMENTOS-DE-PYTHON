# Problema #1: Calculadora de propina inteligente

## Problema
#Solicita total de cuenta y porcentaje de propina. Calcula propina y total final.
#Escriba aquí la descripción del problema.

## Entrada

#- solicitud del monto de la propina
#- solicitud de la propina

## Proceso

#- calcular el monto final de la cuenta a pagar con la propina

## Salida esperada

#- TOTAL FINAL

## Pruebas manuales

| Escenario | Datos de entrada | Resultado esperado | Resultado obtenido |
|-----------|------------------|--------------------|--------------------|
| Caso normal |25000 -15 |28750 |28750 |
| Caso límite | | | |
| Caso inválido | hola| mensaje error| Debe ingresar únicamente valores numéricos. |


```python
# Escriba la solución aquí

try:
    #Entrada
    cuentaTotal = float(input("Ingrese el monto a total de la cuenta a pagar: "))
    propina= float(input("Ingrese el porcentaje(%) de la propina "))
    
    #Proceso
    montoPropina= cuentaTotal*(propina/100)
    totalApagar= cuentaTotal+ montoPropina

    #Salida
    print(f"\nMonto de la cuenta: ₡{cuentaTotal:.2f}")
    print(f"Propina ({propina}%): ₡{montoPropina:.2f}")
    print(f"Total a pagar: ₡{totalApagar:.2f}")
    
except ValueError:
    print("Error: Debe ingresar únicamente valores numéricos.")
    
```

# Problema #2: Detector de número par/impar

## Problema

#Solicita un entero e indica si es par o impar.

## Entrada

#- ingresar numero entero

## Proceso

#- si numero = 0 es par ,sino impar

## Salida esperada

- # el numero ingresado es par o impar

## Pruebas manuales

| Escenario | Datos de entrada | Resultado esperado | Resultado obtenido |
|-----------|------------------|--------------------|--------------------|
| Caso normal |50 |numero par |numero par  |
| Caso límite | 37| numero impar | numero impar|
| Caso inválido |-50 | dato invalido | ingrese  un valor entero positivo |


```python
# Escriba la solución aquí
try:

    #Entrada
    numeroEntero=int(input("Ingrese un numero entero: "))
    
    #Proceso
    if numeroEntero %2 == 0:
        print(f"El numero ingresado {numeroEntero}, es par ")
    else:
        print(f"El numero ingresado {numeroEntero}, es impar ")
        
except ValueError:
    
    print("Por favor ingresar numero entreros positivos ")
```

# Problema #3: Clasificador de edad

## Problema

#Solicita edad y clasifica como menor, adulto o adulto mayor.

## Entrada

- #Solicitud de la edad

## Proceso

- #si la edad es menor a 18 es menor, sino si la edad mayor o igual a 60 es adulto mayor

## Salida esperada

- #menor, adulto o adulto mayor

## Pruebas manuales

| Escenario | Datos de entrada | Resultado esperado | Resultado obtenido |
|-----------|------------------|--------------------|--------------------|
| Caso normal | 65| adulta mayor|adulta mayor |
| Caso límite | 8| amenor|menor  |
| Caso inválido | a | error | ingresar dato numerico |


```python
# Escriba la solución aquí
try:
    #Entrada
    
    edad= int(input("Ingrese la edad de la persona: "))    
    #Proceso
    
    if edad>=60 :
        print(f"La edad ingresada {edad}, es una persona adulta mayor")
    elif edad >=18:
        print(f"La edad ingresada {edad}, es una persona adulta ")
    else:
        print(f"La edad ingresada {edad}, es una persona menor ")
    
except ValueError:
     print("Por favor ingresar datos númericos 😉")
```

# Problema #4: Validador de compra con descuento

## Problema

Solicita cliente, monto y si es cliente frecuente. Aplica descuento según reglas

## Entrada

Nombre del cliente.
Monto de la compra.
Indicar si es cliente frecuente (Sí o No).

## Proceso

Leer los datos ingresados.
Verificar si el cliente es frecuente.
Si es frecuente, aplicar un descuento del 10%.
Calcular el monto del descuento.
Calcular el total a pagar después del descuento.
Mostrar los resultados.

## Salida esperada

Nombre del cliente.
Monto de la compra.
Descuento aplicado.
Total a pagar.

## Pruebas manuales

| Escenario | Datos de entrada | Resultado esperado | Resultado obtenido |
|-----------|------------------|--------------------|--------------------|
| Caso normal |Cliente: Juan, Monto: 50000, Frecuente: Sí |Descuento ₡5000, Total ₡45000 |Correcto |
| Caso límite | Cliente: Ana, Monto: 0, Frecuente: Sí|Descuento ₡0, Total ₡0 | Correcto|
| Caso inválido |Cliente: Pedro, Monto: abc, Frecuente: Sí |Mostrar mensaje de error por monto inválido | Correcto |


```python
# Escriba la solución aquí
#entrada
cliente = input("Ingrese el nombre del cliente: ")
monto = float(input("Ingrese el monto de la compra: "))
frecuente = input("¿Es cliente frecuente? (si/no): ").lower()
#proceso
if frecuente == "si" and monto >= 50000:
    descuento = monto * 0.15
elif frecuente == "si":
    descuento = monto * 0.10
else:
    descuento = 0

total = monto - descuento
#salida
print(f"\nCliente: {cliente}")
print(f"Descuento: ₡{descuento}")
print(f"Total a pagar: ₡{total}")
```
